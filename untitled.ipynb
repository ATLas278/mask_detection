```python
import tensorflow as tf
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import os
import glob

from skimage.io import imread
```

```python
img_dir = os.listdir('data')
for each in img_dir:
    plt.figure()
    currentFolder = 'data/' + each 
    for i, file in enumerate(os.listdir(currentFolder)[0:4]):
        fullpath = currentFolder + '/' + file
        print(fullpath)
        img = imread(fullpath)
        plt.subplot(2,3,i+1)
        plt.imshow(img)
```
